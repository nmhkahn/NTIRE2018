```python
import os
import glob
import numpy as np
import torch
import dataset
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = dataset.TrainDataset("../dataset/DIV2K/x8.h5", "../dataset/DIV2K/gt.h5",
                             "x8", "x4",
                             2, 96)
im_from, im_to = train[10]
im_from = im_from.cpu().mul(255).clamp(0, 255).byte().permute(1, 2, 0).numpy()
im_to   = im_to.cpu().mul(255).clamp(0, 255).byte().permute(1, 2, 0).numpy()


plt.figure()
plt.imshow(im_from)
plt.figure()
plt.imshow(im_to)
```
